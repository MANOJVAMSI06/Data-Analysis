```python

second='https://www.forbesindia.com/article/explainers/top-10-richest-people-india/85909/1'
data=requests.get(second)
data22=BeautifulSoup(data.text,'html')
table=data22.find('table')
table= table.find_all('th')
for th in table:
    print(th.get_text(strip=True))
```

```python
import pandas as pd
#List comperhension
#list Name=[output for loop]
heading=data22.find_all('th')
print(heading)
head=[i.text for i in heading]
print(head)
df=pd.DataFrame(columns=head)
print(df)
```

```python
ind=0
rows=data22.find_all('tr')
for i in rows[1:6]:
   # print(i)
    erd=i.find_all('td')
    #print(erd)
    er=[j.text for j in erd]
    print(er)
    df.loc[ind]=er
    ind+=1
    #print(df)
```
